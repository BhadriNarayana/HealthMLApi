```python
import tensorflow as tf
```

```python
train = tf.keras.utils.image_dataset_from_directory("/kaggle/input/brain-tumor-mri-dataset/Training",
                                                    image_size = (256, 256),
                                                    color_mode = 'rgb',
                                                    label_mode = 'categorical',
                                                    batch_size = 32,
                                                    shuffle = True,
                                                    seed = 12                                                    
)
```

```python
test = tf.keras.utils.image_dataset_from_directory('/kaggle/input/brain-tumor-mri-dataset/Testing',
                                                  image_size = (256, 256),
                                                  color_mode = 'rgb',
                                                  label_mode = 'categorical',
                                                  batch_size = 32,
                                                  seed = 12,
                                                  shuffle = True)
```

```python
tf.random.set_seed(42)
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(20, 5, activation = 'relu'),
    tf.keras.layers.Conv2D(15, 5, activation = 'relu'),
    tf.keras.layers.MaxPooling2D(),
    
    tf.keras.layers.Conv2D(12, 5, activation = 'relu'),
    tf.keras.layers.Conv2D(10, 3, activation = 'relu'),
    tf.keras.layers.MaxPooling2D(),
    
    tf.keras.layers.Conv2D(10, 3, activation = 'relu'),
    tf.keras.layers.MaxPooling2D(),
    
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(4, activation = 'softmax')
    
])
```

```python
model.compile(
    loss = tf.keras.losses.CategoricalCrossentropy(),
    optimizer = tf.keras.optimizers.Adam(),
    metrics = ['accuracy']
)
```

```python
model.fit(train, epochs = 8, steps_per_epoch = len(train), validation_data = test, validation_steps = len(test))
```

```python
model.save('brain_tumor.h5')
```
